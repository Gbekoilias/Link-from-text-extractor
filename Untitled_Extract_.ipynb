```python
#Extracting link from the message column

# Initialize the 'link' column with None
DirtyData['Link'] = None

import re
# Define the URL pattern globally
url_pattern = r'https?://\S+|www\.\S+|bit\.ly/\S+|aka\.ms/\S+'

def extract_link(text):
    # Regular expression pattern to match URLs with various formats
    url_pattern = r'https?://\S+|www\.\S+|bit\.ly/\S+|aka\.ms/\S+'

    # Find all occurrences of the pattern in the text
    matches = re.findall(url_pattern, text)

    # Join multiple matches into a single string separated by ';'
    return ';'.join(matches)
```

```python
DirtyData.head()
```

```python
#Apply Regex function
DirtyData['Link'] = DirtyData['Message'].apply(extract_link)

# Remove the links from the 'message' column
DirtyData['Message'] = DirtyData['Message'].str.replace(url_pattern, '', regex=True)

# Strip any leading or trailing spaces from the 'message' column
DirtyData['Message'] = DirtyData['Message'].str.strip()
```

```python
DirtyData['Message'] = DirtyData['Message'].apply(lambda x: x.rstrip(':'))
```
